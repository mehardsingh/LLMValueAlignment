```python
import pandas as pd
import os
```

```python
dir = "../../analysis_metrics/alignment/"
for filename in os.listdir(dir):
    fp = os.path.join(dir, filename)
    df = pd.read_csv(fp)
    df.loc[df['Section'] == "Overall", 'Section'] = "AllQuestions"

    print(set(df["Section"]))
```

```python
df = pd.read_csv("../../analysis_metrics/alignment/alignment_Llama-2-13b-chat-hf.csv")
df = df[df['Section'] != 'Overall']
df
```

```python
prompts = set(df["Prompt"])
prompts
```

```python
results = list()

for prompt in prompts:
    prompt_df = df[df["Prompt"] == prompt]
    result = prompt_df.groupby('Country')['Alignment'].mean().reset_index()
    result["Section"] = "SectionsAvg"
    result["Prompt"] = prompt
    results.append(result)
    # print(result)

end_res = pd.concat([df] + results, ignore_index=True)
print(end_res)
```
